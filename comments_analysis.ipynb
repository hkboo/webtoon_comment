### **분석 결과가 시사하는 바** 

1. 유저들이 댓글을 통해서 웹툰에 대한 내용 뿐만 아니라 유저들끼리 댓글을 통한 커뮤니케이션과 패러디로 큰 재미를 느낀다는 것이 확인 가능. -> 유명멘트를 패러디, 베스트 댓글 토픽등에서 확인 가능하며 일부 토픽에서는 젠더갈등등 사회적 이슈에 관한 내용들도 볼 수 있었음. -> 유저들끼리의 2차적인 재미를 만들어가고 있으며 유튜브에서 가수 비의 깡 뮤비에 대한 댓글로 2차적인 재미를 만들면서 최근에 이슈화 되는것을 생각하면 이런 2차적인 재미가 웹툰보다 유저들에게 더 큰 재미를 준다고 볼 수 있음. -> **이러한 점을 이용해서 작가입장에서는 댓글에 관한 내용을 작가의 세계관에 반영하는등 유저들과 소통함으로써 유저들에게 또 다른 재미 선사 가능, 네이버 입장에서는 마케팅적으로 '각 주제별로 최고의 댓글을 뽑아라'같은 이벤트를 개최함으로써 댓글에 대한 유저들의 더 큰 호응을 이끌어 낼 수 있음.**

2. 긍정과 부정의 토픽등을 살펴보면 **스토리의 이해여부**가 웹툰의 평가에 큰 영향을 미침을 볼 수 있음.

3. 마지막 장면 토픽을 살펴보면 일반적으로 생각하는 것과 같이 마지막 장면을 어떻게 끝내느냐가 웹툰의 재미에 큰 영향을 끼침을 알 수 있다. 이로 인해 기대감에 관한 댓글도 더욱 생성되며 유저들의 결제를 유도할 수 있다. -> **작가의 입장에서 마지막 장면에 대한 중요성을 확인 할 수 있음**

4. 드라마, 영화제작, 피규어제작에 관한 상당수의 댓글 ->  **유저들의 2차 창작물에 관한 니즈를 파악할 수 있으며 마케팅에 활용이 가능함**.

5. 드라마, 영화제작, '재미가 있는데 순위가 올라가지 않음'에 관한 상당수의 댓글 -> **웹툰을 추천할때의 중요 지표**로 사용이 가능 할 듯.

6. 유저들의 소통을 이끌어낼 수 있는 **웹툰투표, 캐릭터 투표, 댓글의 주제 및 척도 다양화(예를들면 네이트의 찬반댓글), 별점의 주제부여** 등의 마케팅 방안 고려 가능.

```
import re
import tqdm
import random


#전처리가 제대로 안되있는게 존재. 하나의 댓글에 개행이 붙어있음.vi로보니 ^M이라는 문자의 문제 같음. 
def check_comment(comment, reg):  
    
    #개행문자로 인해 정상적이지 않은 댓글 체크
    try:
        check = comment.split(',')[1].isdecimal()
        if not check :
          return False

        comment = "".join(comment.split(',')[2:]).strip()
        #특수문자만 쓰인 댓글들처리
        if not reg.findall(comment):
            return False
        return comment
    except:
        return False
  

#특수문자만 쓰인 댓글처리를 위한 정규식
reg = re.compile('\w+')
#요일마다 존재하는 웹툰이 있어서 중복으로 수집되는 것을 방지
titles_used = set()   
title_info = []
all_random_comments = []
webtoon_comments = []
old_title = '신의 탑 '

with open('./data/naver_webtoon_comments.txt', 'r', encoding='utf-8') as f:
    for comment in tqdm.notebook.tqdm(f):
        now_title = comment.split(',')[0]
        
        #요일마다 존재하는 웹툰이 있어서 중복으로 수집되는 것을 방지
        if now_title in titles_used:
            continue
        
        #댓글에 개행 붙어 있는 경우 + 특수 문자만 쓰인 댓글들 체크
        comment = check_comment(comment,reg)
        if not comment:  
            continue                               
        
        if now_title == old_title :
            webtoon_comments.append(comment)
        else:
            print(now_title)
            try:
                random_comments = random.sample(set(webtoon_comments), 10000)
                all_random_comments.extend(random_comments)
                title_info.extend([old_title]*len(random_comments))
            except:
                all_random_comments.extend(set(webtoon_comments))
                title_info.extend([old_title]*len(set(webtoon_comments)))
            webtoon_comments = [comment]
            titles_used.add(old_title)
            old_title = now_title
```

```
#LDA모델링위한 데이터 준비
from gensim import corpora
from konlpy.tag import Mecab

comments, tokenized_corpus = [], []
tokenizer = Mecab()

for comment in tqdm.notebook.tqdm(all_random_comments):
    tokens = list(set(tokenizer.morphs(comment.strip())))
    comments.append(comment)
    tokenized_corpus.append(tokens)    
```

```
#빈도를 확인 한 후 너무 많거나 적은것들 제거 
from collections import defaultdict
frequency = defaultdict(int)
for text in tokenized_corpus:
    for token in text:
        frequency[token] += 1
```

```
import operator
sorted(frequency.items(), key=operator.itemgetter(1), reverse=True)
```

```
#위의 결과보고 220000초과 제거 
tokenized_corpus = [[token for token in text if frequency[token] < 220000] for text in tokenized_corpus]
tokenized_corpus = [[token for token in text if frequency[token] > 10] for text in tokenized_corpus]
```

```
#key(인덱스) - value(토큰) 쌍으로 매핑.
dictionary = corpora.Dictionary(tokenized_corpus)
#위의 인덱스를 이용해서 각 댓글을 bag of words 형태로 표현함. 
corpus = [dictionary.doc2bow(text) for text in tokenized_corpus]
```

```
#perplexity는 낮을수록 coherence는 높을수록 좋다고 하나 토픽이 커질수록 perplexity는 낮아지고 coherence도 낮아지는 양상을 보임.
#토픽개수별로 어떻게 분류되었는지 보고 최적의 토픽개수를 찾아야 할 듯. 
import pathlib2 
from gensim.models import ldamulticore  
from gensim.models.coherencemodel import CoherenceModel

pathlib2.Path("./lda_model").mkdir(exist_ok=True, parents=True)
num_topics_list = [10,15,20,30,40]
#./lda_model/lda_토픽갯수 형태로 모델들 저장됨.
for num_topics in num_topics_list:
    
    #key(인데스) - value(토큰) 쌍으로 매핑 시켜놓은거랑 bag of words를 데이터로 넣어준다.
    LDA = ldamulticore.LdaMulticore(corpus, id2word=dictionary, num_topics=num_topics,workers=7)
    LDA.save('lda_model/lda_' + str(num_topics))
    print(num_topics)
    print("Perplexity",LDA.log_perplexity(corpus))
    cm = CoherenceModel(model=LDA, corpus=corpus, coherence='u_mass')
    coherence = cm.get_coherence()
    print("Cpherence",coherence)
```

```
from gensim.models import LdaModel
#토픽갯수가 30인 LDA모델 불러옴. 여기서 부터 토픽개수별로 모델 불러와서 최적의 토픽 모델 개수를 찾아야함.
LDA = LdaModel.load('lda_model/lda_30')
all_topics = LDA.get_document_topics(corpus, minimum_probability=0, per_word_topics=False)
```

```
import operator
#각댓글에 해당하는 제일 높은 확률 가진 토픽 뽑아냄. 30개 까지 출력.
for topic, comments in zip(all_topics[:30], all_random_comments[:30]):
    topic = sorted(topic, key = operator.itemgetter(1), reverse = True)
    print(topic[0], comments)
```

```
#결과를 저장 -> 각 토픽에서 높은 확률순으로 댓글 보기 위함 -> 토픽을 해석하기 위한 절차
pathlib2.Path("./result").mkdir(exist_ok=True, parents=True)
with open("./result/lda" + ".results", 'w') as f:
    for doc_idx, topic in tqdm.tqdm_notebook(enumerate(all_topics)):
        topic = sorted(topic, key = operator.itemgetter(1), reverse = True)
        topic_id, prob = topic[0]
        f.writelines(title_info[doc_idx] + "\u241E" + all_random_comments[doc_idx].strip() + "\u241E" + ' '.join(tokenized_corpus[doc_idx]) + "\u241E" + str(topic_id) + "\u241E" + str(prob) + "\n")
```

```
from collections import defaultdict
def load_results(results_fname):
    topic_dict = defaultdict(list)
    with open(results_fname, 'r', encoding='utf-8') as f:
        for line in f:
            title, sentence, _, topic_id, prob = line.strip().split("\u241E")
            topic_dict[int(topic_id)].append((title, sentence, float(prob)))
    for key in topic_dict.keys():
        topic_dict[key] = sorted(topic_dict[key], key=lambda x: x[2], reverse=True)
    return topic_dict
```
